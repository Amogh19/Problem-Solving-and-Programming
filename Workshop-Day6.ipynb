### Groupby
- One of the process to make a set of records to be available as a list according to a condition
- It is not in any means comparable to the filtration or removal of records
- column mentioned in () denotes the column where condition is to be checked
- column mentioned in [] denotes which column is to be printed without any condition checking

```python
df.groupby('Wteam')['Wscore'].mean().head()
```

```python
df.groupby('Wteam')['Wscore'].mean().tail()
```

```python
df.groupby('Wteam')['Season'].mean().head()
```

### Visualizing the data
- An interesting way to display DataFrame
- With matplotlib

```python
import matplotlib.pyplot as plt
a1=[1,2,3,4]
a2=[5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
marks=[79,82,75,63]
divisions=['A','B','C','D']
index=np.arange(len(divisions))
plt.bar(divisions,marks,color="#1412cd",width=0.70)
```

```python
plt.barh(divisions,marks,color='blue')
plt.show()
```

```python
lang=['Python','Java','C','C++']
rating=[75,78,55,65]
plt.pie(rating,labels=lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
